```python
import numpy as np 
import pandas as pd
```

### Creating Data

```python
np.random.seed(1) # For reproducibility
```

```python
no_samples = 1000
class_0_ratio = 0.9
no_class_0 =int(no_samples * class_0_ratio)
no_class_1 = 100
```

```python
class_0_ratio
```

```python
no_class_0
```

```python
no_class_1
```

```python
generated_number_c0 = np.random.normal(0,1, no_class_0)
generated_number_c1 = np.random.normal(3,1,no_class_1)
```

```python
generated_number_c0
```

```python
generated_number_c1
```

### Making Data frame of class 0

```python
[0]* 5
```

```python
class_0 = {'feature1': generated_number_c0,
 'feature2':generated_number_c0,
 'target':[0]*no_class_0}
```

```python
class_0 =  pd.DataFrame(class_0)
```

```python
class_0
```

```python
class1 = pd.DataFrame({'feature1':generated_number_c1,
                       'feature2':generated_number_c1,
                       'target':[1]*no_class_1})
```

```python
class1
```

```python
df = pd.concat([class_0,class1]).reset_index(drop = True)
```

```python
df
```

```python
df.target.value_counts()#(normalize=True)
```

```python
df.target.value_counts(normalize=True)
```

```python
df_minority = df[df.target == 1]
df_majority = df[df.target == 0]
```

```python
df_majority
```

```python
df_minority
```

### Installing Scikit-learn

```python
# UpSampling >> Increasing the minority data to majority number
from sklearn.utils import resample
```

### Resampling | UpSampling

```python
df_minority_upsampled = resample(df_minority, replace=True,  n_samples=len(df_majority), random_state=1)
```

```python
df_minority_upsampled
```

```python
df_minority_upsampled.duplicated().sum()
```

```python
df_upsampled = pd.concat([df_majority,df_minority_upsampled])
```

```python
df_upsampled
```

### DownSampling

```python
df_majority_downsampled = resample(df_majority, replace=False, n_samples=len(df_minority),random_state=1)
```

```python
df_majority_downsampled
```

```python
df_downsampled = pd.concat([df_minority,df_majority_downsampled])
```

```python
df_downsampled
```
