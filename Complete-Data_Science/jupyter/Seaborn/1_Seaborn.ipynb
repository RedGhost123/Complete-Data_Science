### 1(i). Making Histplot of tips datasets

### 1(ii). Making HistPlot of total_bill

```python
sns.histplot(tips['total_bill'])
```

### 1(iii). Making Distplot of total_bill

```python
sns.distplot(tips['total_bill'])
```

### 1(iii[a])

```python
sns.distplot(tips['total_bill'],kde=False)   # KDE = Kernal Density Estimator
```

### 1(iv). Making Scatter plot

```python
sns.scatterplot(x = tips.total_bill,y = tips.tip)
```

```python
tips
```

### 1(v). Making relplot

```python
sns.relplot(x= tips.total_bill,y=tips.tip,data=tips,hue='smoker',style='sex')
```

### 1(vi) Making catplot,barplot and jointPlot

```python
sns.catplot(x='day',y = tips.total_bill,data= tips,color= 'blue')
```

```python
sns.barplot(y=tips.total_bill,x = 'day',data= tips )
```

```python
# catplot
# jointplot
# barplot
sns.jointplot(x = tips.tip,y = tips.total_bill, data = tips)
```

### Making Regression Line in jointplot

```python
sns.jointplot(x = tips.tip,y = tips.total_bill, data = tips,kind="reg")
```

```python
sns.jointplot(x = tips.tip,y = tips.total_bill, data = tips,kind="hex")
```

```python
sns.jointplot(x = tips.tip,y = tips.total_bill, data = tips,kind="kde") # Countor plot maximum used in weather department
```
